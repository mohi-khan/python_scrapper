```python
#This is a web scrapper program to scrape from a blog site
#the site has yearwise listing of blogs in top page
#from each top page you can browse through blog pages

#First import requests file to create the request and beautiful soap for scrapping
import requests
from bs4 import BeautifulSoup

URL = "https://www.uexpress.com/life/dearabby/archives/1991" #This is the URL we are going to scrap
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify()) #To test BS is working
```

```python
blogs=[] #A list of store the blogs
for article in soup.find_all('article'): #In top page blog headlines are listed with article tag
    for row in article.findAll("h2",attrs={'class':'Headline_headline__3vSap Headline_headline_padded__nJGxc'}): #Under article we have h2 tag which store blog heading
        blog={} #Dictionary to store the values of the blog
        blog['header']=row.text #Headline of the blog
        blog['href'] = row.a['href']  #Link of blog page
        URL_link = "https://www.uexpress.com" + blog['href'] #changing relaive link to absolute link
        URL = URL_link
        r = requests.get(URL)
        childsoup = BeautifulSoup(r.content, 'html5lib')
        contents=childsoup.find_all("div",attrs={'class':'Container_container__2GT6K'}) #Div content store all the heading and pages
        my_contents=""
        for content in contents:
            p=content.find("p")
            if p is not None: #It is not a header tag
                my_contents=my_contents + p.text
        blog['contents']=my_contents
        blogs.append(blog)
print(blogs)
```

```python
#Writing on a csv file
import csv
filename = 'blog.csv'
with open(filename, 'w', newline='') as f:
    w = csv.DictWriter(f,['header','href','contents'])
    w.writeheader()
    for blog in blogs:
        w.writerow(blog)
```
